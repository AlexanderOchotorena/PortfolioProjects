```python
#Importing The files and the necessary Library's
import pandas as pd 
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "notebook_connected"


CD = pd.read_excel(r"C:\Users\TheOt\OneDrive\Desktop\PortfolioProjects\Covid-19 Analysis\CovidDeathsForPanda.xlsx")
CV = pd.read_excel(r"C:\Users\TheOt\OneDrive\Desktop\PortfolioProjects\Covid-19 Analysis\CovidVaccinationsForPanda.xlsx")
```

```python
#First We group all the columns by location and continent and sort the values by total deaths DESC to see who had the most deaths
CD1 = CD.groupby(['location','continent']).agg({'total_cases': 'max', 'total_deaths': 'max'}).sort_values(by='total_deaths', ascending=False)
CD1
```

```python
#Finding the leathality percent of each location. 
CD1['leathality_percent'] = (CD1['total_deaths'] / CD1['total_cases']) * 100
CD2 = CD1.round({'leathality_percent' : 3})
CD2
```

```python
#Graphing the data 
#Had to reset the index to allow for the location and continent column to be seen.
CD3 = CD2.reset_index(drop=False)
bar1 = px.bar(
    data_frame=CD3,
    x="location",
    y="leathality_percent",
    color="location",
    title="Leathality percent per location",
    labels={
        "leathality_percent": "Leathality Percentage"
    }
)
bar1
```

```python
#Going to find the total people vaccinated per location 
CDCV = CD.merge(CV)
CDCV1 = CDCV.dropna(subset=['continent']) # Clearing the location values where it is the continent
```

```python
CDCV1['total_vaccinations'] = CDCV1['total_vaccinations'].fillna(0).astype({'total_vaccinations': int}) #Converting the values in only the total_vaccinations column into int and getting rid of the nan values. This fixes an output error.
CDCV1
```

```python
#I grouped everything together by location and continent and got the max values of population and total vac and then sorted by total vac in DESC to see the highest vac rate
CDVC2 = CDCV1.groupby(['location', 'continent']).agg({'total_vaccinations' : 'max', 'population': 'max' }).sort_values(by='total_vaccinations', ascending=False) 
CDVC2
```

```python
#We find the percent of pop vaccinated 
CDVC2['Percent_of_pop_Vaccinated'] = (CDVC2['total_vaccinations'] / CDVC2['population']) * 100
CDVC2.sort_values(by='Percent_of_pop_Vaccinated', ascending=False)
```

```python
#Had to reset the index to allow for the location and continent column to be seen.
CDCV3 = CDVC2.reset_index(drop=False)

#Graphing the data
bar2 = px.bar(
    data_frame=CDCV3,
    x="location",
    y="Percent_of_pop_Vaccinated",
    color="location",
    title= "Percentage Of People Vaccinated Per Location",
    labels={ "Percent_of_pop_Vaccinated": "Percent Of Population Vaccinated"
    }
)
bar2
```
